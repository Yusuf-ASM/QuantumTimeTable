```python
import os
import json
import matplotlib.pyplot as plt
```

```python
path = "./6"
files = os.listdir(path)
files.sort()
temp = files.pop(0)
files.append(temp)
for i in range(len(files)):
    files[i] = path+"/"+files[i]

path = "./16"
files16 = os.listdir(path)
files16.sort()
temp = files16.pop(0)
files16.append(temp)
for i in range(len(files)):
    files16[i] = path+"/"+files16[i]

path = "./24"
files24 = os.listdir(path)
files24.sort()
temp = files24.pop(0)
files24.append(temp)
for i in range(len(files)):
    files24[i] = path+"/"+files24[i]
```

```python
results = []
results16 = []
results24 = []
for i in range(len(files)):
    results.append(json.load(open(str(files[i]))))
    results16.append(json.load(open(str(files16[i]))))
    results24.append(json.load(open(str(files24[i]))))
```

```python
x =[]
y =[]

x16 =[]
y16 =[]

x24 =[]
y24 =[]

x24_20 =[]
y24_20 =[]

for i in results:
    x.append(i['shots'])
    if i['solution_cost'] == 0:
        y.append(i["solution_count"]/i['shots'])
    else:
        y.append(0)

for i in results16:
    x16.append(i['shots'])
    if i['solution_cost'] ==0:
        y16.append(i["solution_count"]/i['shots'])
    else:
        y16.append(0)

for i in results24:
    x24.append(i['shots'])
    if i['solution_cost'] ==0:
        y24.append(i["solution_count"]/i['shots'])
    else:
        y24.append(0)

x,y
```

```python
d={}
d16={}
d24={}

for i in range(len(x)):
    d[x[i]] = y[i]
    d16[x16[i]] = y16[i]
    d24[x24[i]] = y24[i]
```

```python
keyslist = sorted(list(d.keys()))
keyslist16 = sorted(list(d16.keys()))
keyslist24 = sorted(list(d24.keys()))
```

```python
fixedX = keyslist[:]
fixedY = []
for i in keyslist:
    fixedY.append(d[i])

fixedX16 = keyslist16[:]
fixedY16 = []
for i in keyslist16:
    fixedY16.append(d16[i])


fixedX24 = keyslist24[:]
fixedY24 = []
for i in keyslist24:
    fixedY24.append(d24[i])

fixedX,fixedY
```

```python
fixedX16,fixedY16
```

```python
plt.plot(fixedX, fixedY, color='b', label='6Q', ls ="-", marker="o")
plt.xlabel("No. of Shots")
plt.ylabel("Probability of the solution")
plt.legend()
plt.savefig("figs/6Qubits_shots-solution.eps", format='eps')
```
