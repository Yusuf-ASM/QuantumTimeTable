```python
import os
import json
import matplotlib.pyplot as plt
```

```python
save_path = "./new/p_iterations"
shots = [5120, 10240, 20480, 30720]
colors = ["black", "blue", "crimson","teal"] 
x_label = "No. of layers"
y_label = "No. of Minimization Iterations"
```

```python
path = "../results/json/simulation/6"
files = os.listdir(path)
files.sort()
temp = files.pop(0)
files.append(temp)
for i in range(len(files)):
    files[i] = path+"/"+files[i]

path = "../results/json/simulation/16"
files16 = os.listdir(path)
files16.sort()
temp = files16.pop(0)
files16.append(temp)
for i in range(len(files)):
    files16[i] = path+"/"+files16[i]

path = "../results/json/simulation/24"
files24 = os.listdir(path)
files24.sort()
temp = files24.pop(0)
files24.append(temp)
for i in range(len(files)):
    files24[i] = path+"/"+files24[i]

for i in files24:
    print(i)
```

```python
results = []
results16 = []
results24 = []
for i in range(len(files)):
    results.append(json.load(open(str(files[i]))))
    results16.append(json.load(open(str(files16[i]))))
    results24.append(json.load(open(str(files24[i]))))
```

```python
xy6 = {5120: {3: 0, 5: 0, 7: 0, 10: 0}, 10240: {3: 0, 5: 0, 7: 0, 10: 0},
       20480: {3: 0, 5: 0, 7: 0, 10: 0}, 30720: {3: 0, 5: 0, 7: 0, 10: 0}}
xy16 = {5120: {3: 0, 5: 0, 7: 0, 10: 0}, 10240: {3: 0, 5: 0, 7: 0, 10: 0},
        20480: {3: 0, 5: 0, 7: 0, 10: 0}, 30720: {3: 0, 5: 0, 7: 0, 10: 0}}
xy24 = {5120: {3: 0, 5: 0, 7: 0, 10: 0}, 10240: {3: 0, 5: 0, 7: 0, 10: 0},
        20480: {3: 0, 5: 0, 7: 0, 10: 0}, 30720: {3: 0, 5: 0, 7: 0, 10: 0}}


for i in results:
    xy6[i["shots"]][i['layers']] = i['iterations']

for i in results16:
    xy16[i["shots"]][i['layers']] = i['iterations']

for i in results24:
    xy24[i["shots"]][i['layers']] = i['iterations']
```

```python
for shot, color in zip(shots, colors):
    plt.figure()
    x = list(xy6[shot].keys())
    y = list(xy6[shot].values())
    plt.plot(x, y, color=color, label=f'6Q - {shot} shot', ls="-", marker="o")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.savefig(save_path+f"/6_{shot}_p-iteration.eps", format='eps')
    plt.legend()

plt.figure()
for shot, color in zip(shots, colors):
    x = list(xy6[shot].keys())
    y = list(xy6[shot].values())
    plt.plot(x, y, color=color, label=f'6Q - {shot} shot', ls="-", marker="o")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
plt.savefig(save_path+f"/6_all_p-iteration.eps", format='eps')
```
